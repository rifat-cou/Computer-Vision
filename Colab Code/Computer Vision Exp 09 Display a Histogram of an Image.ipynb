```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload image
uploaded = files.upload()
file_name = next(iter(uploaded))
img_data = np.frombuffer(uploaded[file_name], np.uint8)
gray = cv2.imdecode(img_data, cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(6, 3))
plt.imshow(gray)
```

```python
# Step 2: Display the grayscale image
plt.figure(figsize=(6, 3))
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Step 3: Calculate Histogram
hist = cv2.calcHist([gray.astype('float32')], [0], None, [256], [0, 256])
```

```python
# Step 4: Plot Histogram
plt.figure(figsize=(8, 4))
plt.plot(hist, color='black')
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity (0–255)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
