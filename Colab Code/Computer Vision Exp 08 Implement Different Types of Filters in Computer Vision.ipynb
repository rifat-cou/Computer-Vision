```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload grayscale image
uploaded = files.upload()
file_name = next(iter(uploaded))
img_data = np.frombuffer(uploaded[file_name], np.uint8)
gray = cv2.imdecode(img_data, cv2.IMREAD_GRAYSCALE)
plt.imshow(gray)
```

```python
# Step 2: Apply Different Filters

# 2.1 Averaging Filter (Mean)
avg_filtered = cv2.blur(gray, (5, 5))

# 2.2 Gaussian Filter
gaussian_filtered = cv2.GaussianBlur(gray, (5, 5), 0)

# 2.3 Median Filter
median_filtered = cv2.medianBlur(gray, 5)

# 2.4 Laplacian Filter (for sharpening)
laplacian_filtered = cv2.Laplacian(gray, cv2.CV_64F)
laplacian_filtered = cv2.convertScaleAbs(laplacian_filtered)
```

```python
# Step 3: Display Results
titles = ['Original', 'Averaging', 'Gaussian', 'Median', 'Laplacian']
images = [gray, avg_filtered, gaussian_filtered, median_filtered, laplacian_filtered]

plt.figure(figsize=(15, 6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
