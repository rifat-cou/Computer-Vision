```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload image
uploaded = files.upload()
file_name = next(iter(uploaded))
image_data = np.frombuffer(uploaded[file_name], np.uint8)
img = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
plt.imshow(img)
```

```python
# Step 2: Translation (move right by 100, down by 50)
rows, cols = img.shape[:2]
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])
translated_img = cv2.warpAffine(img, translation_matrix, (cols, rows))
```

```python
# Step 3: Rotation (rotate by 45 degrees around the center)
center = (cols // 2, rows // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_img = cv2.warpAffine(img, rotation_matrix, (cols, rows))
```

```python
# Step 4: Scaling (resize to 50% of original size)
scaled_img = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
```

```python
# Step 5: Display images using matplotlib
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (45°)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(scaled_img, cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (50%)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
