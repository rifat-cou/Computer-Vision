```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload image
uploaded = files.upload()
file_name = next(iter(uploaded))
image_data = np.frombuffer(uploaded[file_name], np.uint8)
img = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
plt.imshow(img)
gray = cv2.imdecode(image_data, cv2.IMREAD_GRAYSCALE)
```

```python
# Step 2: Apply Sobel operator
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
```

```python
# Step 3: Apply Laplacian
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
```

```python
# Step 4: Apply Canny edge detector
canny = cv2.Canny(gray, 100, 200)
```

```python
# Step 5: Plot results
plt.figure(figsize=(12, 10))

plt.subplot(2, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(np.absolute(sobel_x), cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(np.absolute(sobel_y), cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(np.absolute(sobel_combined), cmap='gray')
plt.title('Sobel Combined')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(np.absolute(laplacian), cmap='gray')
plt.title('Laplacian')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(canny, cmap='gray')
plt.title('Canny Edge')
plt.axis('off')

plt.tight_layout()
plt.show()
```
