```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload and load grayscale image
uploaded = files.upload()
img_path = next(iter(uploaded))
img_array = np.frombuffer(uploaded[img_path], np.uint8)
img = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
# Step 2: Apply different thresholding techniques
_, thresh_binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
_, thresh_binary_inv = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
_, thresh_trunc = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
_, thresh_tozero = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
_, thresh_tozero_inv = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

#_, is used to store another output from the function which is not needed
# n Python, _ is a convention used for a variable that you don't intend to use.
# In this specific code, the cv2.threshold() function returns two values:
# the threshold value itself and the thresholded image.
# Since the threshold value (127) is already known and not needed for further calculations
# in this particular script, _ is used to signify that we are ignoring that returned value
# and only keeping the second returned value (the thresholded image)
```

```python
# Step 3: Plot all images
titles = ['Original', 'Binary', 'Binary Inv', 'Truncate', 'To Zero', 'To Zero Inv']
images = [img, thresh_binary, thresh_binary_inv, thresh_trunc, thresh_tozero, thresh_tozero_inv]

plt.figure(figsize=(15, 8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Step 4: Print sample pixel values
x, y = 100, 100
print("Original Pixel:", img[y, x])
print("Binary:", thresh_binary[y, x])
print("Binary Inverted:", thresh_binary_inv[y, x])
print("Truncate:", thresh_trunc[y, x])
print("To Zero:", thresh_tozero[y, x])
print("To Zero Inverted:", thresh_tozero_inv[y, x])
```
