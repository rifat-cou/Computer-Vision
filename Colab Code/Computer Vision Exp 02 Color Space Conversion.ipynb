```python
# For Google Colab: Upload and display image using matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload image file
uploaded = files.upload()
file_name = next(iter(uploaded))
img = cv2.imdecode(np.frombuffer(uploaded[file_name], np.uint8), cv2.IMREAD_COLOR)
plt.imshow(img)
```

```python
# Step 2: Convert to different color spaces
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

```python
# Step 3: Display all images using matplotlib
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original (BGR)')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
plt.title('HSV')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(lab, cv2.COLOR_LAB2RGB))
plt.title('LAB')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Step 4: Print pixel value comparisons
x, y = 100, 100  # Choose a pixel
print("BGR Value at (100, 100):", img[y, x])
print("Grayscale Value at (100, 100):", gray[y, x])
print("HSV Value at (100, 100):", hsv[y, x])
print("LAB Value at (100, 100):", lab[y, x])
```
